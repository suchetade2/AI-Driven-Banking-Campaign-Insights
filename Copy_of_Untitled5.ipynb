```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cleaned_bank_marketing_financial_insights.csv")
```

```python
df.head()
df.shape
```

```python
df['subscription_status'].value_counts().plot(kind='bar')
plt.title("Overall Subscription Outcome")
plt.xlabel("Subscription Status")
plt.ylabel("Number of Customers")
plt.show()
```

```python
pd.crosstab(
    df['credit_exposure'],
    df['subscription_status'],
    normalize='index'
).plot(kind='bar', stacked=True)

plt.title("Subscription Outcome by Credit Exposure")
plt.xlabel("Credit Exposure")
plt.ylabel("Proportion")
plt.show()
```

```python
pd.crosstab(
    df['engagement_level'],
    df['subscription_status'],
    normalize='index'
).plot(kind='bar', stacked=True)

plt.title("Subscription Outcome by Engagement Level")
plt.xlabel("Engagement Level")
plt.ylabel("Proportion")
plt.show()
```

```python
pd.crosstab(
    df['campaign_intensity'],
    df['subscription_status'],
    normalize='index'
).plot(kind='bar', stacked=True)

plt.title("Subscription Outcome by Campaign Intensity")
plt.xlabel("Campaign Intensity")
plt.ylabel("Proportion")
plt.show()
```

```python
pd.crosstab(
    df['economic_environment'],
    df['subscription_status'],
    normalize='index'
).plot(kind='bar', stacked=True)

plt.title("Subscription Outcome by Economic Environment")
plt.xlabel("Economic Environment")
plt.ylabel("Proportion")
plt.show()
```

```python
df.boxplot(column='age', by='subscription_status')
plt.title("Age Distribution by Subscription Outcome")
plt.suptitle("")
plt.xlabel("Subscription Status")
plt.ylabel("Age")
plt.show()
```

```python
pd.crosstab(
    df['job'],
    df['subscription_status'],
    normalize='index'
).plot(kind='bar', stacked=True, figsize=(10,5))

plt.title("Subscription Outcome by Job Type")
plt.xlabel("Job")
plt.ylabel("Proportion")
plt.show()
```
